```python
# Use this cell to begin your analyses, and add as many cells as you would like!
import pandas as pd

lego_set = pd.read_csv(PATH_LEGO_SET)
parent_themes = pd.read_csv(PATH_PARENT_THEMES)
```

```python
lego_set.head(5)
```

```python
parent_themes.head(5)
```

```python
merged = lego_set.merge(parent_themes, left_on="parent_theme", right_on="name")

merged.head()
```

```python
merged.drop(columns=["name_y"], inplace=True)

merged.head(5)
```

```python
licenced = merged[merged["is_licensed"]]

licenced.head(5)
```

```python
star_wars = licenced[licenced["parent_theme"]=="Star Wars"]

star_wars.head(5)
```

```python
merged.isna().sum()
```

```python
licenced = licenced.dropna(subset=["set_num"])
```

1. **What percentage of all licensed sets ever released were Star
Wars themed?** Save your answer as a variable `the_force` in the
form of an integer (o.g. 25).

```python
the_force = int((star_wars.shape[0] / licenced.shape[0]) * 100)

print(f"{the_force}%")
```

```python
licenced_sort = licenced.sort_values("year")
licenced_sort["count"] = 1

summed_df = licenced_sort.groupby(["year", "parent_theme"]).sum().reset_index()
```

```python
max_df = summed_df.sort_values("count", ascending=False).drop_duplicates(["year"])
```

```python
max_df.sort_values("year", inplace=True)
max_df
```

2. **In which year was Star Wars not the most popular licensed
theme (in terms of number of sets released that year)?** Save your
answer as a variable `new_era` in the form of an integer (e.g. 2012).

```python
new_era = max_df["year"][-1:]
print(new_era)
```
